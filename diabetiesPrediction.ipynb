import Dependencies


```python
import numpy as np #for create arrays
import pandas as pd #for create DataFrames
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

download dataset using this link:- https://drive.google.com/drive/folders/1Sk1BScxHXbktIbmC0Qk_g6yu8zeGxNr-?usp=drive_link

```python
df = pd.read_csv('/content/drive/MyDrive/DiabetiesPred/diabetes.csv')
```

Seperating data and labels

```python
X = df.drop(columns='Outcome',axis=1)
Y = df['Outcome']
```

```python
print(X,Y)
```

```python
print(Y)
```

Data standartaization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
StandardizedDF = scaler.transform(X)
```

```python
print(StandardizedDF)
```

```python
X = StandardizedDF
Y = df['Outcome']
```

```python
print(X,Y)
```

Train text split

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Train the model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the SVM classifier
classifier.fit(X_train,Y_train)
```

Model evoluation

Accuracy score

```python
x_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(x_train_prediction,Y_train)
```

```python
print("Accuracy score of training data",training_data_accuracy)
```

```python
x_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(x_test_prediction,Y_test)
```

```python
print("Accuracy of the test data",test_data_accuracy)
```

Making a predictive system

```python
input_data =(13,145,82,19,110,22.2,0.245,57)
#Changing the input datato numpyArray

inputDataAsNumpyArray = np.asarray(input_data)
#reshape the array for an instance
inputDataAsNumpyArrayReshaped = inputDataAsNumpyArray.reshape(1,-1)

#Have to standardize the data
std_data = scaler.transform(inputDataAsNumpyArrayReshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)
```
